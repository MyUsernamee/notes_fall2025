$$ \mathcal{L} {f'(t)} = \int_0^\infty e^{-st} f'(t) dt $$ 

$$
\begin{array}{cc}
    u = e^{-st} & v = f(t) \\
    du = -s e^{-st} & dv = f'(t) dt
\end{array} 
$$

$$ 
\begin{align*}
= \left. e^{-st} f(t) \right|_0^{\infty} + \int_0^\infty s e^{-st} f(t) dt \\
= -f(0) + s \mathcal{L} \{f(t)\}
\end{align*}
$$

$$
\mathcal{L} \{f''(t)\} = \int_0^\infty e^{-st} f''(t) dt
$$

$$
\begin{array}{cc}
    u = e^{-st} & v = f'(t) \\
    du = -s e^{-st} & dv = f''(t) dt
\end{array}
$$

$$
\begin{align*}
&= \left. e^{-st} f'(t) \right|_0^\infty + s \int_0^\infty e^{-st} f'(t) dt \\
&= -f'(0) + s [-f(0) + s \mathcal{L} \{f(t)\}] \\
&= -f'(0) - sf(0) + s^2 \mathcal{L} \{f(t)\} \\
&= s^2 F(s) - Sf(0) - f'(0)
\end{align*}
$$
