A heat engine with a power output of 1.80 hp and efficiency 27.0% takes in 684 J of heat each cycle. Find:

            (a) the heat, in Joules, the engine expels each cycle;

            (b) the work, in Joules, the engine does each cycle;

            (c) the time, in seconds, it takes for the engine to complete one cycle.

Note: one horsepower is equal to 746 Watts.

Enter your answer to Part c in the box below.

# Part a)
(a) the heat, in Joules, the engine expels each cycle;

```python
forget()
```

```python
var('Ph Ee Qi')

P_h = Ph == 1.8 * units.power.horsepower
E_e = Ee == 0.27 # %
Q_i = Qi == 684 * units.energy.joule
```

```python
W = Ee * Qi
Qo = Qi - W

Qo_soln = Qo.subs(E_e, Q_i)
Qo_soln
```

# Part b)
(b) the work, in Joules, the engine does each cycle;

```python
W_soln = W.subs(E_e, Q_i)
W_soln
```

# Part c)
(c) the time, in seconds, it takes for the engine to complete one cycle.

```python
# Power = Energy / Time
# Time = Energy / Power

T = W_soln / Ph
T_soln = T.subs(Ph==P_h.rhs().convert(units.power.watt))
T_soln.convert()
```

You fill an ice cube tray with 949 grams of water at a temperature of 21.0 OC and place it in a freezer. The freezer compressor provides power of 45.0 W and the COP of the freezer is 1.80. Find the time, in minutes, that it will take for the water in the tray to become ice at -10.0 OC.

Note: the specific heat capacities of water and ice are 4.186 and 2.093 J/g-OC, respectively, and the latent heat of fusion for water is 333 J/g.

```python
forget()
```

```python
var('Mw Tw Pc COP Tf')

M_w = Mw == 878 * units.mass.gram
T_w = Tw == (21 * units.temperature.celsius)
P_c = Pc == 45 * units.power.watt
COP_ = COP == 1.8
T_f = Tf == (-10 * units.temperature.celsius)
```

```python
C_w = 4.186 * (units.energy.joule / (units.mass.gram * units.temperature.celsius))
C_i = 2.093 * (units.energy.joule / (units.mass.gram * units.temperature.celsius))

C_f = 333 * units.energy.joule / units.mass.gram
```

```python
# Heat needed to get to 0C
Q0 = Mw * (C_w * (-Tw))

Q0_soln = Q0.subs(T_w, M_w)
Q0_soln.convert(units.energy.joule)
```

```python
# Heat needed to fuse to ice

Qf = C_f * Mw

Qf_soln = Qf.subs(M_w)
Qf_soln
```

```python
# Heat to cool ice to -10 C

Q10 = Mw * C_i * (Tf - 0)
Q10_soln = Q10.subs(M_w, T_f)
Q10_soln
```

```python
Q_total = -Q0_soln + Qf_soln - Q10_soln
Q_total
```

```python
# Power = Energy / Time
# Time = Energy / Power

# COP = Qi / W
# COP * W = Qi

Q_required = Q_total / COP_.rhs()

(Q_required / P_c.rhs()).convert(units.time.minute)
```
