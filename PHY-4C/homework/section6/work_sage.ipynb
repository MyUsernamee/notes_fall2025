

On a nice sunny day you take your hot air balloon to a meadow in the highlands, where the air pressure is not one atmosphere.  The balloon and its structure (i.e. the basket, rigging, equipment, etc) has a mass of 270 kg. The air inside the balloon must be heated to a temperature of 145 OC for the balloon to lift off the ground when the outside air temperature is 16.0 OC. To what temperature, in OC, must the air be heated if the balloon is to lift off with three people, each with mass 76.4 kg, people aboard? 

Note: this problem requires three pictures of the balloon (cold, hot, hotter), with the appropriate information labeled in each picture. The algebra is a little tricky... but it is just algebra. Be sure to see the "advice".


```python
m_balloon = 270 * units.mass.kilogram
T_1 = 145 * units.temperature.celsius
T_air = 16 * units.temperature.celsius
m_balloon_people = (270 + 76.4) * units.mass.kilogram
R = 8.314 * units.energy.joule / (units.amount_of_substance.mole * units.temperature.kelvin)
```

```python
T_required = var('T_required')
assume(T_required > 0)
```

```python
P(V, n, T) = n * R * T / V
```

```python
B(V, rho) = rho * V * units.acceleration.gravity
```

```python
F_g_balloon = m_balloon * units.acceleration.gravity
F_g_balloon_people = m_balloon_people * units.acceleration.gravity
F_g_balloon, F_g_balloon_people
```

```python
n_b, V_b, P_b = var('n_b V_b P_b')
rho_a = var('rho_a')
assume(n_b > 0)
assume(V_b > 0)
assume(P_b > 0)
assume(rho_a > 0)
V_b = solve(P_b == P(V_b, n_b, T_1), V_b)[0].rhs()
V_b
```

```python
V_b2, P_b2 = var('V_b2 P_b2')
assume(V_b2 > 0)
assume(P_b2 > 0)
V_b2 = solve(P_b2 == P(V_b2, n_b, T_required), V_b2)[0].rhs()
V_b2
```

```python
B_balloon = B(V_b, rho_a)
B_balloon
```

```python
B_balloon_people = B(V_b2, rho_a)
B_balloon_people
```

```python
n_b = solve(B_balloon - F_g_balloon, [n_b])[0].rhs()
n_b
```

```python
rho(P, T, M) = P * M / (R * T)
rho
```

```python
M_air = var('M_air')
assume(M_air > 0)
rho_a_real = rho(101353 * units.pressure.pascal, T_air, M_air)
rho_a_real
```

```python
n_b = n_b(rho_a=rho_a_real)
n_b
```

```python
solve(B_balloon_people(n_b=n_b, rho_a=rho_a_real) == F_g_balloon_people, [)[0].rhs()
```
