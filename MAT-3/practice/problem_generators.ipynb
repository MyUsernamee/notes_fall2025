## Diagonalization

```python
# For these practice problems, the goal is to generate matrices that have eigen vectors and as such can be diagonalized.
# For this we can start with any generic diagonal matrix which does some form of scaling, then we can rotate, scew, or perform any number of operations that is of course linear,
# and we should be able to recover this original matrix.

from random import randint, randrange, choice

def random_rational(a, b):
    return (Rational(1) / Rational(randint(a, b)) if randint(0, 1) == 0 else randint(a, b)) * choice([-1, 1])
def random_radian():
    return pi / Rational(choice([3, 6, 4]))
    
x_factor = random_rational(1, 4)
y_factor = random_rational(1, 4)
A = Matrix([[x_factor, 0], [0, y_factor]])

# We then can choose random matrix in much the same way
rotation = random_radian()
B = Matrix([[cos(rotation), -sin(rotation)], [sin(rotation), cos(rotation)]])

P = B * A


display(P)
```

```python
lambd = symbols('lambda')
eigen_matrix = eye(2) * lambd

solve(det(P - eigen_matrix),lambd)
```
