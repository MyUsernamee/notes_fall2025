$$ T(\begin{bmatrix} a & b \\ c & d \end{bmatrix}) = a^2 + b^2$$

We already know this is not linear due to the squared terms, which on non-linear.

We can show a proof by contradiction.

$$
\begin{align*}
    T(\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}) &= 1^2 + 0^2 = 1 \\
    T(\begin{bmatrix} 2 & 0 \\ 0 & 0 \end{bmatrix}) &= 2^2 + 0^2 = 4 \\

    T(\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}) + T(\begin{bmatrix} 2 & 0 \\ 0 & 0 \end{bmatrix}) &= 5 \\
    T(\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}) + T(\begin{bmatrix} 2 & 0 \\ 0 & 0 \end{bmatrix}) &\ne T(\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} + \begin{bmatrix} 2 & 0 \\ 0 & 0 \end{bmatrix})
\end{align*}
$$

Therefore $T$ is not linear.
