## New notation

When doing compostion of linear transformation, we introduce a new symbol "$\circ$". When in use it looks like:

$$ f \circ g $$

This is essentially saying, "put g into f" or, "f of g".

$$ f \circ g = g(f(x)) $$

If we think of matricies as "functions" (see 3Blue1Brown's linear algebra series), the intuition for what this means for matricies should make sense.

For matricies, this is just a matrix multiplication:

$$ T_2 \circ T_1 = T_2 T_1 $$
