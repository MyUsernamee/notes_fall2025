```python
%%manim -ql XYExample
from math import *
import numpy as np

class XYExample(Scene):
    def construct(self):
        theta = pi/4
        mat = np.array([[cos(theta), sin(theta)], [-sin(theta), cos(theta)]])
        plane = NumberPlane(background_line_style={
            'stroke_color': GRAY,
            'stroke_opacity': 0.6
        })
        arrow = Vector(plane @ (1, 2))
        prime_plane = NumberPlane()

        xy_mat = Matrix([['\\cos \\frac{\\pi}{2}', '\\sin \\frac{\\pi}{2}'],['-\\sin \\frac{\\pi}{2}', '\\cos \\frac{\\pi}{2}']])

        self.play(Create(xy_mat))
        self.play(xy_mat.animate.move_to(UP * 2 + LEFT * 4), Create(plane), Create(prime_plane))
        self.play( Create(arrow))
        self.wait(1)
        self.play(Indicate(xy_mat), ApplyMatrix(mat, prime_plane))
        self.wait(1)
        self.play(ApplyMatrix(np.linalg.inv(mat), prime_plane), ApplyMatrix(np.linalg.inv(mat), arrow),ApplyMatrix(np.linalg.inv(mat), plane))
        
```
